```python
# Imports
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("../data/raw/environment_tun.csv")

# Preview
df.head()
```

```python
# Remove metadata row
df = df[df["Country Name"] != "#country+name"]

# Reset index
df = df.reset_index(drop=True)

df.head()
```

```python

# Extract Only Relevant Indicators
selected_indicators = [
    "Level of water stress: freshwater withdrawal as a proportion of available freshwater resources",
    "Annual freshwater withdrawals, total (% of internal resources)",
    "Renewable internal freshwater resources, total (billion cubic meters)",
    "Renewable internal freshwater resources per capita (cubic meters)",
    "Agricultural land (% of land area)",
    "Arable land (% of land area)",
    "Average precipitation in depth (mm per year)",
    "Population, total",
    "Forest area (% of land area)"
]

df_selected = df[df["Indicator Name"].isin(selected_indicators)]

df_selected.head()
```

```python
df_pivot = df_selected.pivot(
    index="Year",
    columns="Indicator Name",
    values="Value"
)

df_pivot.head()
```

```python
# Sort by Year
df_pivot = df_pivot.sort_index()
df_pivot.head()
```

```python
# Check for missing values
df_pivot.isnull().sum()
```

```python
# Plot target indicator with safe numeric conversion
import matplotlib.pyplot as plt
target = "Level of water stress: freshwater withdrawal as a proportion of available freshwater resources"

df_target = df[df["Indicator Name"] == target].copy()
# Ensure numeric types for plotting; coerce invalid values to NaN
df_target['Value'] = pd.to_numeric(df_target['Value'], errors='coerce')
df_target['Year'] = pd.to_numeric(df_target['Year'], errors='coerce')
# Drop rows missing the required numeric fields and sort by year
df_target = df_target.dropna(subset=['Value', 'Year']).sort_values('Year')
if df_target.empty:
    print('No numeric data available for the selected indicator.')
else:
    df_target.plot(x='Year', y='Value', figsize=(10,5))
    plt.title('Water Stress Over Time')
    plt.show()
```
